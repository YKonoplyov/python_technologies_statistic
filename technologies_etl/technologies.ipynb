```python
technologies_df = pd.read_json("~/projects/python advanced/python_technologies_statistic/parse_technologies/technologies_data.json")

juniors_df = technologies_df[technologies_df["experience"] < 3]
middle_df = technologies_df[(technologies_df["experience"] > 2) & (technologies_df["experience"] < 5)]
senior_df = technologies_df[technologies_df["experience"] > 4]

technologies_df
```

```python
technologies_all_flat = technologies_df["technologies"].explode()
technologies_jun_flat = juniors_df["technologies"].explode()
technologies_mid_flat = middle_df["technologies"].explode()
technologies_sen_flat = senior_df["technologies"].explode()
```

```python

technologies_df = technologies_df.join(pd.crosstab(technologies_all_flat.index, technologies_all_flat))
juniors_df = juniors_df.join(pd.crosstab(technologies_jun_flat.index, technologies_jun_flat))
middle_df = middle_df.join(pd.crosstab(technologies_mid_flat.index, technologies_mid_flat))
senior_df = senior_df.join(pd.crosstab(technologies_sen_flat.index, technologies_sen_flat))

technologies_df

```

```python
tech_count_all = {technology: technologies_df[technology].sum() for technology in technologies_all_flat if technologies_df[technology].sum() > 1}
tech_count_all = dict(sorted(tech_count_all.items(), key=lambda x:x[1], reverse=True))
tech_count_jun = {technology: juniors_df[technology].sum() for technology in technologies_jun_flat if juniors_df[technology].sum() > 1}
tech_count_jun = dict(sorted(tech_count_jun.items(), key=lambda x:x[1], reverse=True))

tech_count_mid = {technology: middle_df[technology].sum() for technology in technologies_mid_flat if middle_df[technology].sum() > 1}
tech_count_mid = dict(sorted(tech_count_mid.items(), key=lambda x:x[1], reverse=True))

tech_count_sen = {technology: senior_df[technology].sum() for technology in technologies_sen_flat if senior_df[technology].sum() > 1}
tech_count_sen = dict(sorted(tech_count_sen.items(), key=lambda x:x[1], reverse=True))
```

```python
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(40, 20))
plt.bar(tech_count_all.keys(), tech_count_all.values())
plt.xticks(rotation=45)
```

```python
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(40, 20))
plt.bar(tech_count_jun.keys(), tech_count_jun.values())
plt.xticks(rotation=45)
```
